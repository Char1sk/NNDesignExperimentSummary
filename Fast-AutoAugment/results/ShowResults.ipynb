```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def readData(fname):
    with open(fname, 'r') as f:
        lines = f.readlines()
    trainAcc = []
    testAcc = []
    for l in lines:
        trainAcc.append(float(l[34:40]))
        testAcc.append(float(l[65:71]))
        # print(l[34:40])
        # print(l[65:71])
        # break
    # print(trainAcc)
    # print(testAcc)
    return (trainAcc, testAcc)

# readData('./1-1wresnet28x10_cifar_de.txt')
```

```python
def showPlt(trte):
    tr, te = trte
    plt.axis([0, 55, 0.6, 1.0])
    plt.plot(np.linspace(5, 50, 10), tr, label='trainAcc')
    plt.plot(np.linspace(5, 50, 10), te, label='testAcc')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
```

```python
tr11, te11 = readData('./1-1wresnet28x10_cifar_de.txt')
tr12, te12 = readData('./1-2wresnet28x10_cifar_aa.txt')
tr13, te13 = readData('./1-3wresnet28x10_cifar_fa.txt')
tr21, te21 = readData('./2-1wresnet4x2_cifar_de.txt')
tr22, te22 = readData('./2-2wresnet4x2_cifar_aa.txt')
tr23, te23 = readData('./2-3wresnet4x2_cifar_fa.txt')
tr41, te41 = readData('./4-1wresnet28x10_cifar_de_200.txt')
tr42, te42 = readData('./4-2wresnet28x10_cifar_fa_200.txt')
```
