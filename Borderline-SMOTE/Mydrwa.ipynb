```python
import matplotlib.pyplot as plt
```

```python
with open('./result1.txt') as f:
    lines = f.readlines()
    trainacc1 = []
    testacc1 = []
    for l in lines:
        if 'Train' in l and 'Acc' in l:
            a = l.find('Acc ')
            b = l.find('%')
            i = float(l[a+4:b])*100
            trainacc1.append(i)
        if 'Test' in l and 'Acc' in l:
            a = l.find('Acc ')
            b = l.find('%')
            i = float(l[a+4:b])*100
            testacc1.append(i)
with open('./result2.txt') as f:
    lines = f.readlines()
    trainacc2 = []
    testacc2 = []
    for l in lines:
        if 'Train' in l and 'Acc' in l:
            a = l.find('Acc ')
            b = l.find('%')
            i = float(l[a+4:b])*100
            trainacc2.append(i)
        if 'Test' in l and 'Acc' in l:
            a = l.find('Acc ')
            b = l.find('%')
            i = float(l[a+4:b])*100
            testacc2.append(i)
with open('./result3.txt') as f:
    lines = f.readlines()
    trainacc3 = []
    testacc3 = []
    for l in lines:
        if 'Train' in l and 'Acc' in l:
            a = l.find('Acc ')
            b = l.find('%')
            i = float(l[a+4:b])*100
            trainacc3.append(i)
        if 'Test' in l and 'Acc' in l:
            a = l.find('Acc ')
            b = l.find('%')
            i = float(l[a+4:b])*100
            testacc3.append(i)
with open('./result4.txt') as f:
    lines = f.readlines()
    trainacc4 = []
    testacc4 = []
    for l in lines:
        if 'Train' in l and 'Acc' in l:
            a = l.find('Acc ')
            b = l.find('%')
            i = float(l[a+4:b])*100
            trainacc4.append(i)
        if 'Test' in l and 'Acc' in l:
            a = l.find('Acc ')
            b = l.find('%')
            i = float(l[a+4:b])*100
            testacc4.append(i)
```

```python
plt.plot(testacc1, '-')
plt.plot(testacc2, '-')
plt.plot(testacc3, '-')
plt.plot(testacc4, '-')
plt.legend(['exp1', 'exp2', 'exp3', 'exp4'])
plt.show()
```
